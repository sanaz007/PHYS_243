#### Let's start coding with creating a random walk simulation!

```python
import numpy as np
import random
```

```python
def walk_2d(steps):
    x, y = 0, 0
    walks = [[x, y]]
    for _ in range(int(steps)):
        choose_rand = random.choice([(-1, 0), (1, 0), (0, -1), (0, 1)])
        dx, dy = choose_rand
        x, y = x + dx, y + dy
        walks.append([x, y])
    return walks
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8, 8))

for i in range(5):
    run = np.array(walk_2d(10**4)).T
    plt.plot(run[0], run[1], label=f"random walk {i}")

plt.legend()
plt.show()
```
